```python
import sys
import os
sys.path.append(os.path.abspath("../src"))
from preprocessing import load_and_prepare
from model import train_model, save_model
from evaluate import evaluate_model

BASE_DIR = os.path.abspath("..")
phish_path = os.path.join(BASE_DIR, "data", "phishing_emails.csv")
legit_path = os.path.join(BASE_DIR, "data", "legit_emails.csv")

print("Working directory:", os.getcwd())
print("Files in data/:", os.listdir(os.path.join(os.getcwd(), "..", "data")))

X_train, X_test, y_train, y_test = load_and_prepare(phish_path, legit_path)

model = train_model(X_train, y_train)
evaluate_model(model, X_test, y_test)
model_path = os.path.join(BASE_DIR, "models", "phishnet_model.pkl")

print("Model path:", model_path)
print("Directory exists:", os.path.isdir(os.path.dirname(model_path)))
save_model(model, path=model_path)

import os
print(os.path.exists("models/phishnet_model.pkl"))
```
